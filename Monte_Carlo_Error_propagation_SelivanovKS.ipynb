```python
import numpy as np

%matplotlib notebook

import matplotlib.pyplot as plt
plt.xkcd()
```

```python
#Part 1
M1 = 40 * 10**4
M1_err = 0.05 * 10**4
M2 = 30 * 10**4
M2_err = 0.1 * 10**4
r = 3.2
r_err = 0.01
G = 6.67384 * 10**(-11)
```

```python
F = G * M1 * M2 / r**2
F_err = np.sqrt((G * M2 / r**2)**2 * M1_err**2 + (G * M1 / r**2)**2 * M2_err**2 + (-2 * G * M1 * M2 / r**3)**2 * r_err**2)

print('F =', F, '+/-', F_err)
```

```python
N = 1000000
M1_m = np.random.normal(M1, M1_err, N)
M2_m = np.random.normal(M2, M2_err, N)
r_m = np.random.normal(r, r_err, N)
F_m = G * M1_m * M2_m / r_m ** 2

print('F =', np.mean(F_m), '+/-', np.std(F_m))
```

```python
x = np.linspace(0.76, 0.81, 1000)
y = 1 / (F_err * np.sqrt(2 * np.pi)) *  np.exp(-(x - F)**2 / (2 * F_err**2))

plt.hist(F_m, bins=50, range=[0.76, 0.81], density=True)
plt.plot(x, y)
plt.show()
```
